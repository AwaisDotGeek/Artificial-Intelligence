```python
# -1 shows empty location

goalMatrix = [[1,2,3],
              [4,5,6],
              [7,8,-1]]

startMatrix = [[1,2,3],
              [-1,4,6],
              [7,5,8]]

def printMatrix(mat):
    for i in range(len(mat)):
        for j in range(len(mat)):
            print(mat[i][j], end = ' ')
        print()
    print()
        
printMatrix(goalMatrix)
print()
printMatrix(startMatrix)
```

```python
def isGoalStateAchieved(goalState, currentState):
    if goalState == currentState:
        return True

def getHeuristicValue(goalMat, currentMat):
    heuristic = 0
    
    for i in range(len(goalMat)):
        for j in range(len(currentMat)):
            if goalMat[i][j] != currentMat[i][j]:
                heuristic += 1
    
    return heuristic
            
def moveSpecifiedIndexInMatrix(matrix, row, col, direction): 
    if direction == 'l':
        matrix[row][col] = matrix[row][col - 1]
        matrix[row][col - 1] = -1
        
    elif direction == 'u':
        matrix[row][col] = matrix[row - 1][col]
        matrix[row - 1][col] = -1
        
    elif direction == 'd':
        matrix[row][col] = matrix[row + 1][col]
        matrix[row + 1][col] = -1
        
    elif direction == 'r':
        matrix[row][col] = matrix[row][col + 1]
        matrix[row][col + 1] = -1
        
def getEmptyValueIndex(matrix):
    rowCol = []
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if matrix[i][j] == -1:
                rowCol.append(i)
                rowCol.append(j)
                break
    return rowCol
    
def getMatWithLeastHearisticValue(heauristicValues, matrices):
    print('\nBranches to:')
    
    minIndex = 0
    for i in range(len(heauristicValues)):
        if heauristicValues[i] < heauristicValues[minIndex]:
            minIndex = i
        
        printMatrix(matrices[i])
    
    return matrices[minIndex]
    
def getCopyOfMatrix(matrix):
    copyOfMatrix = []
    for i in range(len(matrix)):
        copyOfMatrix.append([])
        for j in range(len(matrix)):
            copyOfMatrix[i].append(matrix[i][j])
        
    return copyOfMatrix

def reachGoalState(goalMat, currentMat):
    rowColOfEmptyLocation = getEmptyValueIndex(currentMat)
    row = rowColOfEmptyLocation[0]
    col = rowColOfEmptyLocation[1]
    
    heuristicValues = []
    matrices = []
    
    # create copy of current Matrix
    copyOfCurrentMat = getCopyOfMatrix(currentMat)
    
    print('<----- Current Matrix ----->')
    printMatrix(currentMat)
    
    if row == 0:
        if col == 1:            
            # Move right
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveRight(currentMat, row, col)
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)
            
            # Move Down
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'd')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)
            
            # Move left
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'l')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)

            currentMat = getMatWithLeastHearisticValue(heuristicValues, matrices)
            
        elif col == 0:            
            # Move right
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'r')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)
            
            # Move Down
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'd')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)

            currentMat = getMatWithLeastHearisticValue(heuristicValues, matrices)
            
            print("Current State")
            printMatrix(currentMat)
            print()
            
        else:            
            # Move Down
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'd')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)
            
            # Move left
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'l')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)

            currentMat = getMatWithLeastHearisticValue(heuristicValues, matrices)
            
    elif row == 1:
        if col == 0:
            # Move up
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'u')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)
            
            # Move right
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'r')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)
            
            # Move Down
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'd')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)

            currentMat = getMatWithLeastHearisticValue(heuristicValues, matrices)
            
        elif col == 1:
            # Move up
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'u')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)
            
            # Move right
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'r')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)
            
            # Move Down
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'd')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)
            
            # Move left
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'l')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)

            currentMat = getMatWithLeastHearisticValue(heuristicValues, matrices)
            
        else:
            # Move up
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'u')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)
            
            # Move Down
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'd')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)
            
            # Move left
            currentMat = getCopyOfMatrix(copyOfCurrentMat)
            moveSpecifiedIndexInMatrix(currentMat, row, col, 'l')
            heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
            matrices.append(currentMat)

            currentMat = getMatWithLeastHearisticValue(heuristicValues, matrices)
            
    elif row == 2:
            if col== 0:
                # Move up
                moveSpecifiedIndexInMatrix(currentMat, row, col, 'u')
                heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
                matrices.append(currentMat)

                # Move right
                currentMat = getCopyOfMatrix(copyOfCurrentMat)
                moveSpecifiedIndexInMatrix(currentMat, row, col, 'r')
                heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
                matrices.append(currentMat)

                currentMat = getMatWithLeastHearisticValue(heuristicValues, matrices)

            elif col == 1:
                # Move up
                moveSpecifiedIndexInMatrix(currentMat, row, col, 'u')
                heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
                matrices.append(currentMat)

                # Move right
                currentMat = getCopyOfMatrix(copyOfCurrentMat)
                moveSpecifiedIndexInMatrix(currentMat, row, col, 'r')
                heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
                matrices.append(currentMat)

                # Move left
                currentMat = getCopyOfMatrix(copyOfCurrentMat)
                moveSpecifiedIndexInMatrix(currentMat, row, col, 'l')
                heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
                matrices.append(currentMat)

                currentMat = getMatWithLeastHearisticValue(heuristicValues, matrices)

            else:
                # Move up
                moveSpecifiedIndexInMatrix(currentMat, row, col, 'u')
                heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
                matrices.append(currentMat)

                # Move left
                currentMat = getCopyOfMatrix(copyOfCurrentMat)
                moveSpecifiedIndexInMatrix(currentMat, row, col, 'l')
                heuristicValues.append(getHeuristicValue(goalMatrix, currentMat))
                matrices.append(currentMat)

                currentMat = getMatWithLeastHearisticValue(heuristicValues, matrices)
                
    if(not isGoalStateAchieved(goalMatrix, currentMat)):
        reachGoalState(goalMatrix, currentMat)
    else:
        print("Goal achieved sucksesfuli") 
        printMatrix(currentMat)
```
